LCDM CLASS parameters:

{
    'Omega_b': 0.0491989, 
    'h': 0.6737, 
    'n_s': 0.96605, 
    'ln10^{10}A_s': 3.0447, 
    'Omega_m': 0.3145714273, 
    'output': 'mPk', 
    'non linear': 'halofit', 
    'z_max_pk': 2.5, 
    'P_k_max_1/Mpc': 30.0
}

Then, the suppression is applied by calling the CWDM emulator with m = 50 eV, f = 0.2.

```python
import numpy as np
from matplotlib import pyplot as plt

loaded_file = np.load("pk_lcdm_cwdm.npz")
ks = loaded_file['ks']

plt.figure(figsize=(8,8))
plt.xscale("log")
plt.xlim(ks[0], ks[-1])
plt.yscale("log")
plt.xlabel("k [1/Mpc]")
plt.ylabel("P(k)")

plt.plot(
    ks,
    loaded_file['pk_lcdm_z1'],
    label="LCDM z=0.5",
    color="black",
)
plt.plot(
    ks,
    loaded_file['pk_lcdm_z2'],
    label="LCDM z=1",
    color="black",
    linestyle="dashed"
)
plt.plot(
    ks,
    loaded_file['pk_cwdm_z1'],
    label="CWDM z=0.5",
    color="red",
)
plt.plot(
    ks,
    loaded_file['pk_cwdm_z2'],
    label="CWDM z=1",
    color="red",
    linestyle="dashed"
)

plt.legend()
```

```python
plt.figure(figsize=(8,8))
plt.xscale("log")
plt.xlim(ks[0], ks[-1])
plt.title("m = 50 eV, f = 0.2")
plt.xlabel("k [1/Mpc]")
plt.ylabel("P(k)_CWDM / P(k)_LCDM")

plt.plot(
    ks,
   loaded_file['pk_cwdm_z1'] / loaded_file['pk_lcdm_z1'],
    label="CWDM / LCDM z=0.5",
    color="black"
)
plt.plot(
    ks,
    loaded_file['pk_cwdm_z2'] / loaded_file['pk_lcdm_z2'],
    label="CWDM / LCDM z=1",
    color="black",
    linestyle="dashed"
)

plt.legend()
```
